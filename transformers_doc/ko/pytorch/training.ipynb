ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì„ ë¯¸ì„¸ íŠœë‹í•˜ê¸° ìœ„í•´ì„œ ë°ì´í„°ì…‹ì„ ë‹¤ìš´ë¡œë“œí•˜ê³  í›ˆë ¨í•  ìˆ˜ ìˆë„ë¡ ì¤€ë¹„í•˜ì„¸ìš”. ì´ì „ íŠœí† ë¦¬ì–¼ì—ì„œ í›ˆë ¨ì„ ìœ„í•´ ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ëŠ” ë°©ë²•ì„ ë³´ì—¬ë“œë ¸ëŠ”ë°, ì§€ê¸ˆì´ ë°°ìš¸ ê±¸ ë˜ì§šì„ ê¸°íšŒì…ë‹ˆë‹¤!

ë¨¼ì € [Yelp ë¦¬ë·°](https://huggingface.co/datasets/yelp_review_full) ë°ì´í„° ì„¸íŠ¸ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤:

````python
from datasets import load_dataset

dataset = load_dataset("yelp_review_full")
dataset["train"][100]
````

í…ìŠ¤íŠ¸ë¥¼ ì²˜ë¦¬í•˜ê³  ì„œë¡œ ë‹¤ë¥¸ ê¸¸ì´ì˜ ì‹œí€€ìŠ¤ íŒ¨ë”© ë° ì˜ë¼ë‚´ê¸° ì „ëµì„ í¬í•¨í•˜ë ¤ë©´ í† í¬ë‚˜ì´ì €ê°€ í•„ìš”í•©ë‹ˆë‹¤. ë°ì´í„°ì…‹ì„ í•œ ë²ˆì— ì²˜ë¦¬í•˜ë ¤ë©´ ğŸ¤— Dataset [`map`](https://huggingface.co/docs/datasets/process#map) ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì „ì²´ ë°ì´í„°ì…‹ì— ì „ì²˜ë¦¬ í•¨ìˆ˜ë¥¼ ì ìš©í•˜ì„¸ìš”:

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-cased")


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)


tokenized_datasets = dataset.map(tokenize_function, batched=True)
````

í•„ìš”í•œ ê²½ìš° ë¯¸ì„¸ íŠœë‹ì„ ìœ„í•´ ë°ì´í„°ì…‹ì˜ ì‘ì€ ë¶€ë¶„ ì§‘í•©ì„ ë§Œë“¤ì–´ ë¯¸ì„¸ íŠœë‹ ì‘ì—… ì‹œê°„ì„ ì¤„ì¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤:

ğŸ¤— TransformersëŠ” ğŸ¤— Transformers ëª¨ë¸ í›ˆë ¨ì— ìµœì í™”ëœ [Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer) í´ë˜ìŠ¤ë¥¼ ì œê³µí•˜ì—¬ í›ˆë ¨ ë£¨í”„ë¥¼ ì§ì ‘ ì‘ì„±í•˜ì§€ ì•Šê³ ë„ ì‰½ê²Œ í›ˆë ¨ì„ ì‹œì‘í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. [Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer) APIëŠ” ë¡œê¹…(logging), ê²½ì‚¬ ëˆ„ì (gradient accumulation), í˜¼í•© ì •ë°€ë„(mixed precision) ë“± ë‹¤ì–‘í•œ í›ˆë ¨ ì˜µì…˜ê³¼ ê¸°ëŠ¥ì„ ì§€ì›í•©ë‹ˆë‹¤.

### í•˜ì´í¼íŒŒë¼ë¯¸í„° í›ˆë ¨[[training-hyperparameters]]

ë‹¤ìŒìœ¼ë¡œ ì •í•  ìˆ˜ ìˆëŠ” ëª¨ë“  í•˜ì´í¼íŒŒë¼ë¯¸í„°ì™€ ë‹¤ì–‘í•œ í›ˆë ¨ ì˜µì…˜ì„ í™œì„±í™”í•˜ê¸° ìœ„í•œ í”Œë˜ê·¸ë¥¼ í¬í•¨í•˜ëŠ” [TrainingArguments](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.TrainingArguments) í´ë˜ìŠ¤ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

ì´ íŠœí† ë¦¬ì–¼ì—ì„œëŠ” ê¸°ë³¸ í›ˆë ¨ [í•˜ì´í¼íŒŒë¼ë¯¸í„°](https://huggingface.co/docs/transformers/main_classes/trainer#transformers.TrainingArguments)ë¡œ ì‹œì‘í•˜ì§€ë§Œ, ììœ ë¡­ê²Œ ì‹¤í—˜í•˜ì—¬ ì—¬ëŸ¬ë¶„ë“¤ì—ê²Œ ë§ëŠ” ìµœì ì˜ ì„¤ì •ì„ ì°¾ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

í›ˆë ¨ì—ì„œ ì²´í¬í¬ì¸íŠ¸(checkpoints)ë¥¼ ì €ì¥í•  ìœ„ì¹˜ë¥¼ ì§€ì •í•©ë‹ˆë‹¤:

````python
from transformers import TrainingArguments

training_args = TrainingArguments(output_dir="test_trainer")
````

### í‰ê°€ í•˜ê¸°[[evaluate]]

[Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer)ëŠ” í›ˆë ¨ ì¤‘ì— ëª¨ë¸ ì„±ëŠ¥ì„ ìë™ìœ¼ë¡œ í‰ê°€í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. í‰ê°€ ì§€í‘œë¥¼ ê³„ì‚°í•˜ê³  ë³´ê³ í•  í•¨ìˆ˜ë¥¼ [Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer)ì— ì „ë‹¬í•´ì•¼ í•©ë‹ˆë‹¤.
[ğŸ¤— Evaluate](https://huggingface.co/docs/evaluate/index) ë¼ì´ë¸ŒëŸ¬ë¦¬ëŠ” [`evaluate.load`](https://huggingface.co/spaces/evaluate-metric/accuracy) í•¨ìˆ˜ë¡œ ë¡œë“œí•  ìˆ˜ ìˆëŠ” ê°„ë‹¨í•œ `accuracy`í•¨ìˆ˜ë¥¼ ì œê³µí•©ë‹ˆë‹¤ (ìì„¸í•œ ë‚´ìš©ì€ [ë‘˜ëŸ¬ë³´ê¸°](https://huggingface.co/docs/evaluate/a_quick_tour)ë¥¼ ì°¸ì¡°í•˜ì„¸ìš”):

````python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
````

`metric`ì—ì„œ `compute`ë¥¼ í˜¸ì¶œí•˜ì—¬ ì˜ˆì¸¡ì˜ ì •í™•ë„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. ì˜ˆì¸¡ì„ `compute`ì— ì „ë‹¬í•˜ê¸° ì „ì— ì˜ˆì¸¡ì„ ë¡œì§“ìœ¼ë¡œ ë³€í™˜í•´ì•¼ í•©ë‹ˆë‹¤(ëª¨ë“  ğŸ¤— Transformers ëª¨ë¸ì€ ë¡œì§“ìœ¼ë¡œ ë°˜í™˜í•œë‹¤ëŠ” ì ì„ ê¸°ì–µí•˜ì„¸ìš”):

````python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
````

ë¯¸ì„¸ íŠœë‹ ì¤‘ì— í‰ê°€ ì§€í‘œë¥¼ ëª¨ë‹ˆí„°ë§í•˜ë ¤ë©´ í›ˆë ¨ ì¸ìˆ˜ì— `eval_strategy` íŒŒë¼ë¯¸í„°ë¥¼ ì§€ì •í•˜ì—¬ ê° ì—í­ì´ ëë‚  ë•Œ í‰ê°€ ì§€í‘œë¥¼ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤:

````python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", eval_strategy="epoch")
````

### í›ˆë ¨ í•˜ê¸°[[trainer]]

ëª¨ë¸, í›ˆë ¨ ì¸ìˆ˜, í›ˆë ¨ ë° í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹, í‰ê°€ í•¨ìˆ˜ê°€ í¬í•¨ëœ [Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer) ê°ì²´ë¥¼ ë§Œë“­ë‹ˆë‹¤:

````python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
````

ê·¸ë¦¬ê³  [train()](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer.train)ì„ í˜¸ì¶œí•˜ì—¬ ëª¨ë¸ì„ ë¯¸ì„¸ íŠœë‹í•©ë‹ˆë‹¤:

````python
trainer.train()
````

### DataLoader[[dataloader]]

í›ˆë ¨ ë° í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì— ëŒ€í•œ 'DataLoader'ë¥¼ ìƒì„±í•˜ì—¬ ë°ì´í„° ë°°ì¹˜ë¥¼ ë°˜ë³µí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤:

````python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(small_train_dataset, shuffle=True, batch_size=8)
eval_dataloader = DataLoader(small_eval_dataset, batch_size=8)
````

ì˜ˆì¸¡ì„ ìœ„í•œ ë ˆì´ë¸” ê°œìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì„ ë¡œë“œí•©ë‹ˆë‹¤:

````python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-cased", num_labels=5)
````

### ì˜µí‹°ë§ˆì´ì € ë° í•™ìŠµ ì†ë„ ìŠ¤ì¼€ì¤„ëŸ¬[[optimizer-and-learning-rate-scheduler]]

ì˜µí‹°ë§ˆì´ì €ì™€ í•™ìŠµ ì†ë„ ìŠ¤ì¼€ì¤„ëŸ¬ë¥¼ ìƒì„±í•˜ì—¬ ëª¨ë¸ì„ ë¯¸ì„¸ ì¡°ì •í•©ë‹ˆë‹¤. íŒŒì´í† ì¹˜ì—ì„œ ì œê³µí•˜ëŠ” [`AdamW`](https://pytorch.org/docs/stable/generated/torch.optim.AdamW.html) ì˜µí‹°ë§ˆì´ì €ë¥¼ ì‚¬ìš©í•´ ë³´ê² ìŠµë‹ˆë‹¤:

````python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
````

[Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer)ì—ì„œ ê¸°ë³¸ í•™ìŠµ ì†ë„ ìŠ¤ì¼€ì¤„ëŸ¬ë¥¼ ìƒì„±í•©ë‹ˆë‹¤:

````python
from transformers import get_scheduler

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    name="linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps
)
````

ë§ˆì§€ë§‰ìœ¼ë¡œ, GPUì— ì•¡ì„¸ìŠ¤í•  ìˆ˜ ìˆëŠ” ê²½ìš° 'device'ë¥¼ ì§€ì •í•˜ì—¬ GPUë¥¼ ì‚¬ìš©í•˜ë„ë¡ í•©ë‹ˆë‹¤. ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ CPUì—ì„œ í›ˆë ¨í•˜ë©° ëª‡ ë¶„ì´ ì•„ë‹Œ ëª‡ ì‹œê°„ì´ ê±¸ë¦´ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### í›ˆë ¨ ë£¨í”„[[training-loop]]

í›ˆë ¨ ì§„í–‰ ìƒí™©ì„ ì¶”ì í•˜ë ¤ë©´ [tqdm](https://tqdm.github.io/) ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ íŠ¸ë ˆì´ë‹ ë‹¨ê³„ ìˆ˜ì— ì§„í–‰ë¥  í‘œì‹œì¤„ì„ ì¶”ê°€í•˜ì„¸ìš”:

````python
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
````

### í‰ê°€ í•˜ê¸°[[evaluate]]

[Trainer](https://huggingface.co/docs/transformers/main/ko/main_classes/trainer#transformers.Trainer)ì— í‰ê°€ í•¨ìˆ˜ë¥¼ ì¶”ê°€í•œ ë°©ë²•ê³¼ ë§ˆì°¬ê°€ì§€ë¡œ, í›ˆë ¨ ë£¨í”„ë¥¼ ì§ì ‘ ì‘ì„±í•  ë•Œë„ ë™ì¼í•œ ì‘ì—…ì„ ìˆ˜í–‰í•´ì•¼ í•©ë‹ˆë‹¤. í•˜ì§€ë§Œ ì´ë²ˆì—ëŠ” ê° ì—í¬í¬ê°€ ëë‚  ë•Œë§ˆë‹¤ í‰ê°€ì§€í‘œë¥¼ ê³„ì‚°í•˜ì—¬ ë³´ê³ í•˜ëŠ” ëŒ€ì‹ , `add_batch`ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë“  ë°°ì¹˜ë¥¼ ëˆ„ì í•˜ê³  ë§¨ ë§ˆì§€ë§‰ì— í‰ê°€ì§€í‘œë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

````python
import evaluate

metric = evaluate.load("accuracy")
model.eval()
for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)
    metric.add_batch(predictions=predictions, references=batch["labels"])

metric.compute()
````
